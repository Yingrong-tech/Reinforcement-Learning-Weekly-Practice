## 2.2 Default Arguments and Keyword Arguments

Functions can have default arguments and can be invoked using keyword arguments for better flexibility.

1. Define a function named `power` that takes two arguments, `base` and `exponent`, and returns the result of `base` raised to the power of `exponent`. The `exponent` argument should have a default value of 2:

```python
# create a function and add a default argument
def power(base, exponent=2):
    return base ** exponent

# Invoke the function
result = power(3)  # Default exponent value is used
print("Result:", result)

result = power(2, exponent=3)  # Custom exponent value is used
print("Result:", result)
```

2. Define a function named `calculate` that takes two numbers as arguments and performs various arithmetic operations. Use keyword arguments to selectively perform the operations:

```python
# create a function 'calculate'
def calculate(a, b, add=False, subtract=False, multiply=False, divide=False):
    results = []

    if add:
        results.append(a + b)
    if subtract:
        results.append(a - b)
    if multiply:
        results.append(a * b)
    if divide:
        results.append(a / b)

    return results

# Invoke the function
results = calculate(5, 3, subtract=True, divide=True)
print("Results:", results)
```

```python
# Rewritting the above function to use match--case (Similar to Switch--case in other languages) Available from Python 3.10+
def calculate(a, b, operation): # operation can be 'add', 'sub', 'mul, 'div'
    results = []
    for option in operation:
      match option:
        case 'add':
          results.append(a + b)
        case 'sub':
          results.append(a - b)
        case 'mul':
          results.append(a * b)
        case 'div':
          results.append(a / b)
        case _:
          results.append('Unknown operation')

    return results

# Invoke the function
results = calculate(5, 3, operation=['sub', 'div', 'hi'])
print("Results:", results)
```

### Plot multiple line together in the same plot and add axis labels, title, and legend

```python
# Create a sine curve
# Find the coordinates of the curve for plotting
xcoord = np.arange(0, 2 * np.pi, 0.1)
ycoord = np.sin(xcoord)

y_sin = np.sin(xcoord) # Sine Curve
y_cos = np.cos(xcoord) # Cos curve

# Plot the curve
plt.plot(xcoord, y_sin)
plt.plot(xcoord, y_cos)

# Add title
plt.title('Sine and Cos curves')

# Add axis labels
plt.xlabel('X axis')
plt.ylabel('y axis')

# Add legend
plt.legend(['Sin', 'Cos'])

# Show the origin line y=0
plt.axhline(y=0, c="red", label="y=0")

# How can you add a vertical lnne from origin? i.e. x=0?
```

### 2. Explore Subplots

Subplots are used to plot multiple curve/result in the same figure.

```python
# Create a sine curve
# Find the coordinates of the curve for plotting
xcoord = np.arange(0, 4 * np.pi, 0.1)
ycoord = np.sin(xcoord)

y_sin = np.sin(xcoord) # Sine Curve
y_cos = np.cos(xcoord) # Cos curve

# Create subplot grid with:
# Rows = 2
# Cols = 1
plt.subplot(2,1,1) # The 3rd paramenter indicates the location of the plot in the grid

# Plot the 1st curve
plt.plot(xcoord, y_sin, c="red")
plt.title('Sine Curve')
#Show the origin line y=0
plt.axhline(y=0, c="blue", label="y=0")

# Setup and Plot the 2nd curve
plt.subplot(2,1,2)
plt.plot(xcoord, y_cos, c="green")
plt.title('Cos Curve')
#Show the origin line y=0
plt.axhline(y=0, c="blue", label="y=0")

# Add title
plt.title('Sine and Cos curves')

# Show the plot
plt.show()
```
